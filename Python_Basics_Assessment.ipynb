```python
#3 Compare and contrast mutable and immutable objects in Python with examples.
#Ans:Mutable Objects: Mutable objects are those whose state or content can be changed after they are created.
1.	Lists: You can change, add, or remove elements.
    My_list = [1, 2, 3]
    My_list.append(4)  # my_list is now [1, 2, 3, 4]
    My_list[0] = 0     # my_list is now [0, 2, 3, 4]
2.	Dictionaries: You can change values, add new key-value pairs, or remove existing ones.
    My_dict = {‘a’: 1, ‘b’: 2}
    My_dict[‘c’] = 3    # my_dict is now {‘a’: 1, ‘b’: 2, ‘c’: 3}
    My_dict[‘a’] = 0    # my_dict is now {‘a’: 0, ‘b’: 2, ‘c’: 3}
3.	Sets: You can add or remove elements. 
    My_set = {1, 2, 3}
    My_set.add(4)       # my_set is now {1, 2, 3, 4}
    My_set.remove(2)    # my_set is now {1, 3, 4}
Immutable Objects: immutable objects are those whose state or content cannot be changed once they are created.
1.	Strings:Any operation that seems to modify a string actually creates a new string.
    My_string = “hello”
    New_string = my_string.replace(‘h’, ‘j’)  # my_string is still “hello”; new_string is “jello”
2.	Tuples: Elements cannot be added, removed, or changed.
    My_tuple = (1, 2, 3)
3.	Integers and Floats: Numbers are immutable.Q
    My_int = 10
    My_int += 1   # my_int is now 11, but the original integer 10 remains unchanged somewhere in memory.

Differences between mutable and immutable objects:
Mutable can be modified after creation whereas immutable cannot be modified after creation.
Mutable generally have more memory overhead due to their ability to change whereas immutable often more memory-efficient since they don’t change, allowing for certain optimizations (like interning).
Mutable are useful when you need to maintain a collection of items that can change, such as a list of users, a dictionary of settings, etc., whereas Immutable are useful for constants, ensuring data integrity (since they can’t be changed inadvertently), and as keys in dictionaries because they are hashable.
```

```python
#4 Discuss the different types of operators in Python and provide examples of how they are used.
#Ans:The different types of operators are:
 1. Arithmetic Operators
Perform basic mathematical operations.
Addition (+):
    a = 5 + 3  # a is 8
Subtraction (-):
    b = 5 – 3  # b is 2
Multiplication (*):
    c = 5 * 3  # c is 15
Division (/):
    d = 5 / 3  # d is 1.666…
 
Modulus (%): (remainder)
    f = 5 % 3  # f is 2
2. Comparison Operators – Compare two values.
Equal to (==):
    a == 5  # True if a is 5
Not equal to (!=):
    a != 5  # True if a is not 5
Greater than (>):   
    a> 3  # True if a is greater than 3
Less than (<):
    a< 3  # True if a is less than 3
Greater than or equal to (>=):
    a>= 3  # True if a is greater than or equal to 3
Less than or equal to (<=):
    a <= 3  # True if a is less than or equal to 3

3.Assignment Operators – Assign values to variables.

Assignment (=):
    a = 5  # a is 5
Addvc and assign (+=):
    a += 3  # a is now 8 (a = a + 3)
Subtract and assign (-=):
    a -= 3  # a is now 5 (a = a – 3)
Multiply and assign (*=):
    a *= 3  # a is now 15 (a = a * 3)

Divide and assign (/=):
   a /= 3  # a is now 5.0 (a = a / 3)

4.	Logical Operators: Used to combine conditional statements.

AND (and):
    a > 3 and a < 10  # True if both conditions are True
OR (or):
    a > 3 or a < 2  # True if at least one condition is True   
NOT (not):
    Not(a > 3)  # True if a is not greater than 3
```

```python
#7.Describe the different types of loops in Python and their use cases with examples.
#Ans:Python has two main types of loops: ‘for’ loops and ‘while’ loops. These loops allow you to repeat a block of code multiple times.
1.for Loop
Repeats a block of code for each item in a sequence (like a list, tuple, string, or range).
Use Case:  Use a for loop when you know the number of iterations in advance.
Examples:
Iterating over a list:
    Fruits = [“apple”, “banana”, “cherry”]
    For fruit in fruits:
        Print(fruit)

Output:
    Apple
    Banana
    Cherry

Using `range` to iterate over numbers:
    For I in range(5):
        Print(i)

Output:
0
1
2
3
4

2.while Loop
Repeats a block of code as long as a condition is true.
Use Case: Use a `while` loop when you want to repeat something until a condition changes.
Examples:
Counting down:
    Count = 5
    While count > 0:
        Print(count)
        Count -= 1

Output:
    5
    4
    3
    2
    1

3.break: Exits the loop immediately.
    For I in range(5):
        If i == 3:
            break
        print(i)
    
Output:
    0
    1
    2

4.continue:Skips the current iteration and moves to the next one.
    for i in range(5):
        if i == 3:
            continue
        print(i)
   
Output:
    0
    1
    2
    4
```
